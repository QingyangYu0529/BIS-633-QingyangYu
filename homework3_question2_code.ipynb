```python
# import library pandas
import pandas as pd
data = pd.read_csv("BRFSS_19_data.csv")
```

```python
# The FIPS code of Connecticut is 09.
# Selected the data to include only respondents in the state of Connecticut.
data_CT = data[data['_STATE']==9]
```

```python
data_CT
```

```python
# Chose the exercise Jogging(code = 28), selected all the people that spend the most time to jog.
data_jogging = data[data['EXRACT11']==28.0]
data_jogging
```

```python
# Using groupby to group all the people that favourite jogging by gender.
data_jogging_by_sex = data_jogging.groupby('_SEX')[['Unnamed: 0']].count()
data_jogging_by_sex
```

```python
# Removed all the 'NaN' from dataframe data_CT, and saved into the list data_CT_height. 
data_CT_height = data_CT[data_CT['HTIN4'].notna()]
# Extracted the column 'HTIN4', saved into the list data_CT_height.
data_CT_height = data_CT_height['HTIN4']
data_CT_height
```

```python
# Calculated the average height(unit: inch).
avg_height_inch = sum(data_CT_height)/len(data_CT_height)
# Transferred the unit into meters.
avg_height_meter = avg_height_inch * 0.0254
# Print the final result of average height.
print(f"The average height for a person in the state Connecticut is {avg_height_meter:.2f} meters.")
```
